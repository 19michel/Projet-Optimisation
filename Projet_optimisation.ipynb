# Trajectoire optimale de course d'un véhicule électrique

Dans ce sujet, on considère la course d'un véhicule électrique sur une piste. On cherche à optimiser le profil de vitesse du véhicule afin qu'il parcoure le nombre maximal de tours de piste pour une puissance électrique initiale donnée. On connait le plan exact de la piste à parcourir. La vitesse est supposée bornée supérieurement et inférieurement.

## 1 Modélisation

1) On peut raisonnablement supposer que le véhicule va parcourir un grand nombre de tours. Dans ce cas, on peut même ajouter que ce nombre sera entier car une partie de tour devient alors negligeable devant le grand nombre qui vont être réalisés. 

Ainsi, il est raisonnable de dire que tous les tours seront identiques car la quantité d'energie nécessaire pour réaliser un tour est negligeable devant celle totale qui permet de réaliser un grand nombre de tours. Dans ce cas, pour réaliser un nombre maximal de tours, il suffit de minimiser l'énergie dépensée pour faire avancer le véhicule pendant un tour de piste.

En effet, si on note $N$ le nombre de tours réalisés, $E$ l'énergie totale et $e$ l'énergie dépensé pour un tour, on a, avec les hypothèses précédentes:

$$ N = \frac{E}{e}$$

Donc, si on minimise $e$, on arrive à maximiser $N$.

----------

2) Calculons l'expression de l'énergie $e$ dépensée pendant un tour d'une durée $t$.

$$ e = \int_{0}^{t} \mathcal{P} dt $$

$$ e = \int_{0}^{t} U_{m} I_{m} dt $$

$$ e = \int_{0}^{t} (I_{m} R_{m} + \kappa \omega) I_{m} dt $$

$$ e = \int_{0}^{t} \left( \frac{R_{m}}{\kappa^{2}} \mathcal{T}^{2} + \omega \mathcal{T} \right) dt $$

Or on a $\omega = \frac{U_{m}}{\kappa} - \frac{R_{m}}{\kappa^{2}} \mathcal{T}$, d'où

$$ e = \int_{0}^{t} \frac{U_{m}}{\kappa} \mathcal{T}  dt $$

On introduit ensuite l'abscisse curviligne X qui permet de se repérer sur le circuit. Le cricuit est alors caractérisé par un arc paramétré $\mathcal{A}$ qui à une abscisse curviligne associe une position dans l'espace.

$$ \mathcal{A} : \begin{split}
[0,L] \to \mathbb{R}^{3} \\
X \mapsto (x,y,z)
\end{split}$$

Puis, en introduisant l'abscisse curviligne dans l'intégrale à l'aide d'un changement de variable et de la définition de la vitesse $V = \frac{dX}{dt}$, on obtient

$$ e = \int_{0}^{L} \frac{U_{m}}{\kappa} \mathcal{T}  \frac{dX}{V} $$

Fnalement, pour simplifier, on pose $\mathcal{T}(X) = \tau u(X)$ où $\tau$ est une constante et $u$ est une fonction variant entre $0$ et $1$. Cela revient en réalité à contrôler le couple pour piloter le véhicule. En effet, la fonction $u$ traduit le contrôle effectué par l'utilisateur du circuit électrique.

$$ \fbox{$ e = \int_{0}^{L} \frac{U_{m} \tau}{\kappa V} u  dX $} $$

***

3) On s'intéresse ici à une *roue du véhicule*.

Au niveau de I, le point de contact, en supposant qu'il y ait **roulement sans glissement**, on a $v_{I}^{sol} = v_{I}^{roue}$.

Or, on sait que $v_{I}^{sol} = V$ où $V$ représente la vitesse du véhicule.

D'autre part, on connait $\omega$, la vitesse de rotation de la roue et $R$ son rayon, donc on a $v_{I}^{roue} = \omega R$.

Finalement, on obtient que $ V = \omega R $.

Puis, en complexifiant le modèle avec l'ajout d'un **rapport de transmission** $\eta$, on a $v_{I}^{sol} = \eta v_{I}^{roue}$.

On obtient donc la relation finale suivante

$$ \fbox{$V = \eta \omega R$} $$

***

4) Dans cette question, on commence par effectuer un **bilan des forces** s'appliquant sur la voiture.

Les différentes forces qui interviennent sont:

-la *réaction du support* $4\vec{R_{n}}$ qui ne nous intéressent pas ici

-le *poids* $\vec{P}$ qui se projète sur l'axe du circuit, c'est-à-dire l'axe dirigé par $\frac{\vec{dX}}{dX}$, comme $-mg sin(\theta)$.

-la *force de traction* $\vec{F_{t}}$ qui vaut sur l'axe du circuit $\mathcal{T}R$.

-la *force de frottements linéaires* $\vec{F_{f}}$ qui vaut sur l'axe du circuit $-\alpha V$.

Ainsi, en appliquant le **principe fondamental de la dynamique** projeté sur l'axe du circuit, on a

$$ \fbox{$m\frac{dV}{dt} = \mathcal{T}R - \alpha V - mg sin(\theta)$} $$

***

5) En reprenant la question précédente, on a

$$ m\frac{dV}{dX}\frac{dX}{dt} = \tau R u - \alpha V - mg sin(\theta) $$

$$ mV\frac{dV}{dX} = \tau R u - \alpha V - mg sin(\theta) $$

On introduit alors $n$ points de passage équidistants sur le plan de course $X_{0}, \dots, X_{n-1}$ tels que $X_{i} = \frac{i}{n}L$ puis on discrétise la fonction $V$.

On a, avec $V_{k}=V(X_{k})$ $$ mV_{k}n\frac{V_{k+1} - V_{k}}{L} = \tau R u(X_{k}) - \alpha V_{k} - mg sin(\theta(X_{k})) $$

$$ V_{k+1} = V_{k} + \frac{L}{mnV_{k}} \left( \tau R u(X_{k}) - \alpha V_{k} - mg sin(\theta(X_{k})) \right) $$

$$ \fbox{$V_{k+1} = V_{k} + \frac{L}{mnV_{k}} \left( \tau R u(X_{k}) - \alpha V_{k} - mg sin(\frac{dX}{dz}(X_{k})) \right)$} $$

On connait $\frac{dX}{dz}(X_{k})$ car on connait le tracé du circuit. Ainsi, on peut exprimer $V$ en fonction de $u$ la fonction de contrôle.

Mais, il faut minimiser $e$. On peut d'ailleurs discrétiser l'expression de cette énergie en approximant l'intégrale par une somme.

$$ \fbox{$ e = \frac{L}{n} \sum_{k=0}^{L} \frac{U_{m} \tau}{\kappa V_{k}} u(X_{k}) $} $$

Finalement, à l'aide d'une condition initiale, on cherche

$$ \fbox{$ \min_{U \in \mathbb{R}^{n}} e(U)$} $$

où $U = (u(X_{0}), \dots, u(X_{n-1}))$
